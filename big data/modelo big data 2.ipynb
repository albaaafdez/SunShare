```python
import pandas as pd
file_path = "C:\\Users\\gonza\\Documents\\Universidad Europea\\Tercero\\Big Data I\\energia_madrid_actualizado.csv"
data = pd.read_csv(file_path)
data
```

## Consumo total por tipo de edificio, media de consumo y cantidad de edificios de cada

```python
# Calcular el consumo total y la cantidad de edificios por tipo de edificio
consumo_total_por_tipo_edificio = data.groupby('tipo_del_edificio')['consumo_kWh'].agg(['sum', 'mean', 'count']).reset_index()

# Renombrar las columnas
consumo_total_por_tipo_edificio.columns = ['tipo de edificio', 'Consumo Total (kWh)', 'Media de Consumo (kWh)', 'Cantidad de Edificios']

# Ordenar por consumo total de mayor a menor
consumo_total_por_tipo_edificio = consumo_total_por_tipo_edificio.sort_values(by='Media de Consumo (kWh)', ascending=False)

# Mostrar el resultado
consumo_total_por_tipo_edificio
```

```python
data_ordenada = data.groupby(['edificio', 'agno', 'mes'])['consumo_kWh'].sum().reset_index()
data_ordenada = data_ordenada.sort_values(by=['edificio', 'agno', 'mes']).reset_index(drop=True)
data_ordenada
```

## Buscamos los edificios que tienen registros los ultimos 3 años

```python
data_ultimos_3_años = data_ordenada[data_ordenada['agno'].isin([2022, 2023, 2024])]

data_suma_anual = data_ultimos_3_años.groupby(['edificio', 'agno'])['consumo_kWh'].sum().reset_index()

edificios_consumo_positivo = (
    data_suma_anual[data_suma_anual['consumo_kWh'] > 0]
    .groupby('edificio')['agno']
    .count()
)

edificios_consumo_positivo

edificios_consumo_positivo_3_años_list = edificios_consumo_positivo[edificios_consumo_positivo == 3].index.tolist()

tabla_edificios_consumo_positivo = data_suma_anual[data_suma_anual['edificio'].isin(edificios_consumo_positivo_3_años_list)]

tabla_edificios_consumo_positivo
```

## Entrenamos el modelo Random Forest

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

edificios_unicos = tabla_edificios_consumo_positivo['edificio'].unique()

# Crear una lista para almacenar las predicciones por edificio para el año 2025
predicciones_por_edificio_2025 = []

# Iterar sobre cada edificio y entrenar un modelo específico con Random Forest
for edificio in edificios_unicos:
    # Filtrar los datos para el edificio actual
    data_edificio = tabla_edificios_consumo_positivo[tabla_edificios_consumo_positivo['edificio'] == edificio]
    
    # Crear X (año) y y (consumo_kWh) para el edificio actual
    X_edificio = data_edificio[['agno']]
    y_edificio = data_edificio['consumo_kWh']
    
    # Dividir los datos en entrenamiento (2022-2024)
    X_train_edificio = X_edificio[X_edificio['agno'] <= 2024]
    y_train_edificio = y_edificio[X_edificio['agno'] <= 2024]
    
    # Crear y entrenar el modelo de Random Forest
    modelo_edificio_rf = RandomForestRegressor(random_state=42)
    modelo_edificio_rf.fit(X_train_edificio, y_train_edificio)
    
    # Realizar predicción para 2025
    y_pred_edificio_2025 = modelo_edificio_rf.predict([[2025]])
    
    # Almacenar las predicciones
    predicciones_por_edificio_2025.append({
        'Edificio': edificio,
        'Año': 2025,
        'Consumo Predicho (kWh)': max(0, y_pred_edificio_2025[0])  # Evitar valores negativos
    })

# Convertir la lista de predicciones a un DataFrame
predicciones_2025_df = pd.DataFrame(predicciones_por_edificio_2025)

# Guardar las predicciones del año 2025 en un archivo CSV
predicciones_2025_df.to_csv('Predicciones_Consumo_2025.csv', index=False)

# Confirmación
print("Archivo guardado como 'Predicciones_Consumo_2025.csv'")
```

## Resultados del modelo 

```python
file_path = "C:\\Users\\gonza\\Documents\\Universidad Europea\\Tercero\\Big Data I\\Predicciones_Consumo_2025.csv"
resultados_modelo = pd.read_csv(file_path)
resultados_modelo
```

## Ordenamos los resultados por tipo de edificio

```python
resultados_modelo.columns = resultados_modelo.columns.str.lower().str.strip()

# Realizar la unión
predicciones_rf_final = resultados_modelo.merge(
    data[['edificio', 'tipo_del_edificio']].drop_duplicates(),
    on='edificio',
    how='left'
)

predicciones_rf_final = predicciones_rf_final.sort_values(by='tipo_del_edificio')

predicciones_rf_final
```

## Cantidad de edificios de cada tipo de los que se han usado para entrenar el modelo

```python
conteo_tipos_edificio_df = predicciones_rf_final['tipo_del_edificio'].value_counts().reset_index()
conteo_tipos_edificio_df.columns = ['Tipo de Edificio', 'Cantidad']
conteo_tipos_edificio_df
```

## Media de los consumos predichos ordenados por tipo de edificio de mayor a menor

```python
media_consumo_tipo_edificio = predicciones_rf_final.groupby('tipo_del_edificio')['consumo predicho (kwh)'].mean().reset_index()
media_consumo_tipo_edificio.columns = ['Tipo de Edificio', 'Media de Consumo Predicho (kWh)']
media_consumo_tipo_edificio = media_consumo_tipo_edificio.sort_values(by='Media de Consumo Predicho (kWh)', ascending=False)
media_consumo_tipo_edificio
```
