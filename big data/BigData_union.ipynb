```python
import pandas as pd

# Intentar con la codificación correcta
consumo = pd.read_csv('consumo_energia.csv', encoding='utf-8', sep=',')
Inventario = pd.read_csv('inventario_completo_limpio.csv', encoding='utf-8', sep=',')

consumo.columns = consumo.columns.str.lower()
Inventario.columns = Inventario.columns.str.lower()

# Mostrar las primeras filas para verificar
consumo


```

```python
#  Ver cuántos valores nulos hay en cada columna
print("Número de valores nulos por columna:")
print(Inventario.isnull().sum())

```

```python
#  Ver cuántos valores nulos hay en cada columna
print("Número de valores nulos por columna:")
print(consumo.isnull().sum())

```

```python
DISTRITO = consumo.groupby(['distrito']).count()
conteo_dist_consumo = DISTRITO.sort_values(by=['edificio'], ascending=True)
conteo_dist_consumo
```

```python
DISTRITO = Inventario.groupby(['distrito']).count()
conteo_dist_inventario = DISTRITO.sort_values(by=['centro'], ascending=True)
conteo_dist_inventario
```

```python
# Eliminar columnas innecesarias
columnas_a_eliminar = ['número', 'código ayto', 'coordenada-x', 'coordenada-y', 'latitud', 'longitud','adscripción', 'superficie (m2)']
Inventario = Inventario.drop(columns=columnas_a_eliminar)

# Verificar el resultado
Inventario.head()
```

```python
# Cambiar el nombre de las columnas
nuevos_nombres = {
    'centro': 'edificio',
    'uso del edificio': 'tipo del edificio'
}

Inventario = Inventario.rename(columns=nuevos_nombres)

# Verificar el cambio
Inventario.head()
```

```python
# Guardar el dataset modificado en un nuevo archivo CSV
Inventario.to_csv('inventario_modificado.csv', index=False, encoding='utf-8')
```
